```python
lista_moedas = [100, 100, 25, 25, 25, 10, 1, 1, 1, 1]
valor = 10

def Problema_do_Troco(lista_moedas, valor):
    n = len(lista_moedas)
    iteracoes = 0

    # dp[i][v] = True se é possível formar o valor v usando as primeiras i moedas
    dp = [[False for _ in range(valor + 1)] for _ in range(n + 1)]

    # para reconstruir quais moedas foram escolhidas
    escolha = [[False for _ in range(valor + 1)] for _ in range(n + 1)]

    # valor 0 sempre é possível: não escolher nada
    for i in range(n + 1):
        dp[i][0] = True

    for i in range(1, n + 1):
        moeda = lista_moedas[i - 1]
        for v in range(valor + 1):
            iteracoes += 1

            # não pegar a moeda
            dp[i][v] = dp[i - 1][v]

            # tentar pegar a moeda
            if v >= moeda and dp[i - 1][v - moeda]:
                dp[i][v] = True
                escolha[i][v] = True

    if not dp[n][valor]:
        moedas = []
    else:
        moedas = []
        v = valor
        i = n

        while v > 0 and i > 0:
            if escolha[i][v]:
                moedas.append(lista_moedas[i - 1])
                v -= lista_moedas[i - 1]
            i -= 1

        moedas.reverse()

    return {
        "Moedas selecionados": moedas,
        "valor": valor,
        "iteracoes": iteracoes,
    }

resultado = Problema_do_Troco(lista_moedas, valor)
print(resultado)
```
