# Atividade: 

1. **N-rainhas — primeira solução** No pior caso, o algoritmo tenta posicionar rainhas em várias colunas para cada linha. Complexiade = O(n!)

Isso acontece porque, a cada linha, o algoritmo escolhe uma coluna válida, evitando repetir colunas. No pior caso, ele pode testar muitas permutações possíveis.
A complexidade de espaço é = a O(n) (uma rainha por linha)

2. **N-rainhas — todas as soluções** O algoritmo vai alem da primeira resposta. Complexidade = O(n!)

Porém, na prática, é mais custoso que a primeira solução, porque percorre todas as possibilidades válidas e inválidas até terminar. A complexidade de espaço é: O(n + s * n) --> s = número de soluções encontradas

--> pilha da recursão, ele guarda todas as soluções.


3. **Soma dos subconjuntos — primeira solução** pergunta é incluir ou não incluir --> Pior caso = O(2^n) | A complexidade de espaço = O(n)

4. **Soma dos subconjuntos — todas as soluções** testar todos os subconjuntos possíveis --> O(2^n) | A complexidade de espaço = O(n + s * n) --> s = número de subconjuntos encontrados cuja soma é zero

```python
import time
import random
import pandas as pd

def n_rainhas_primeira(n): # n rainhas primeira solucao 
    sol = []
    colunas = set()
    diag1 = set()
    diag2 = set()

    cont = {"chamadas": 0, "iteracoes": 0}

    def backtracking(linha):
        cont["chamadas"] += 1

        if linha == n:
            return True

        for coluna in range(n):
            cont["iteracoes"] += 1

            if coluna in colunas or linha - coluna in diag1 or linha + coluna in diag2:
                continue

            sol.append(coluna)
            colunas.add(coluna)
            diag1.add(linha - coluna)
            diag2.add(linha + coluna)

            if backtracking(linha + 1):
                return True

            sol.pop()
            colunas.remove(coluna)
            diag1.remove(linha - coluna)
            diag2.remove(linha + coluna)

        return False

    inicio = time.time()
    achou = backtracking(0)
    fim = time.time()

    if achou:
        return sol, cont, fim - inicio
    return None, cont, fim - inicio


def n_rainhas_todas(n): # n rainhas toadas as solucoes 
    solucoes = []
    sol = []
    colunas = set()
    diag1 = set()
    diag2 = set()

    cont = {"chamadas": 0, "iteracoes": 0}

    def backtracking(linha):
        cont["chamadas"] += 1

        if linha == n:
            solucoes.append(sol.copy())
            return

        for coluna in range(n):
            cont["iteracoes"] += 1

            if coluna in colunas or linha - coluna in diag1 or linha + coluna in diag2:
                continue

            sol.append(coluna)
            colunas.add(coluna)
            diag1.add(linha - coluna)
            diag2.add(linha + coluna)

            backtracking(linha + 1)

            sol.pop()
            colunas.remove(coluna)
            diag1.remove(linha - coluna)
            diag2.remove(linha + coluna)

    inicio = time.time()
    backtracking(0)
    fim = time.time()

    return solucoes, cont, fim - inicio


def mostrar_tabuleiro(sol): # mostrar tabuleiros 
    if sol is None:
        print("Sem solução")
        return

    n = len(sol)

    for coluna_rainha in sol:
        linha = ""

        for coluna in range(n):
            if coluna == coluna_rainha:
                linha += "Q "
            else:
                linha += ". "

        print(linha)


def soma_subconjunto_primeira(vetor): # soma --> primeiara solucao 
    sol = []
    n = len(vetor)

    cont = {"chamadas": 0, "iteracoes": 0}

    def backtracking(i, soma):
        cont["chamadas"] += 1

        if soma == 0 and len(sol) > 0:
            return True

        if i == n:
            return False

        cont["iteracoes"] += 1

        # Escolhe o elemento
        sol.append(vetor[i])

        if backtracking(i + 1, soma + vetor[i]):
            return True

        sol.pop()

        # Não escolhe o elemento
        if backtracking(i + 1, soma):
            return True

        return False

    inicio = time.time()
    achou = backtracking(0, 0)
    fim = time.time()

    if achou:
        return sol.copy(), cont, fim - inicio

    return None, cont, fim - inicio


def soma_subconjunto_todas(vetor): # todas as solucoes 
    solucoes = []
    sol = []
    n = len(vetor)

    cont = {"chamadas": 0, "iteracoes": 0}

    def backtracking(i, soma):
        cont["chamadas"] += 1

        if i == n:
            if soma == 0 and len(sol) > 0:
                solucoes.append(sol.copy())
            return

        cont["iteracoes"] += 1

        sol.append(vetor[i]) # --> esolhe elemento
        backtracking(i + 1, soma + vetor[i])

        sol.pop()

        backtracking(i + 1, soma) # nao escolhe 

    inicio = time.time()
    backtracking(0, 0)
    fim = time.time()

    return solucoes, cont, fim - inicio

def gerar_conjunto(tamanho, minimo=-100, maximo=100): # gera conjunto aleatorio 
    vetor = []

    for i in range(tamanho):
        num = random.randint(minimo, maximo)

        while num == 0:
            num = random.randint(minimo, maximo)

        vetor.append(num)

    return vetor

# resultados 

def tabela_resultados():
    resultados = []

    # testes
    for n in [4, 5, 6, 7, 8]:
        sol, cont, tempo = n_rainhas_primeira(n)

        resultados.append({
            "problema": "N-rainhas",
            "tipo": "Primeira solução",
            "entrada": f"n = {n}",
            "qtd_solucoes": 1 if sol else 0,
            "chamadas": cont["chamadas"],
            "iteracoes": cont["iteracoes"],
            "tempo": tempo
        })

        sols, cont, tempo = n_rainhas_todas(n)

        resultados.append({
            "problema": "N-rainhas",
            "tipo": "Todas as soluções",
            "entrada": f"n = {n}",
            "qtd_solucoes": len(sols),
            "chamadas": cont["chamadas"],
            "iteracoes": cont["iteracoes"],
            "tempo": tempo
        })

    # Testes soma dos subconjuntos
    testes = [
        [-7, -3, -2, 5, 8],
        [1, 2, 3, 4, 5, 10],
        [-5, 2, 3, -1, 1],
        gerar_conjunto(20),
        gerar_conjunto(25)
    ]

    for vetor in testes:
        sol, cont, tempo = soma_subconjunto_primeira(vetor)

        resultados.append({
            "problema": "Soma dos subconjuntos",
            "tipo": "Primeira solução",
            "entrada": f"tamanho = {len(vetor)}",
            "qtd_solucoes": 1 if sol else 0,
            "chamadas": cont["chamadas"],
            "iteracoes": cont["iteracoes"],
            "tempo": tempo
        })

        sols, cont, tempo = soma_subconjunto_todas(vetor)

        resultados.append({
            "problema": "Soma dos subconjuntos",
            "tipo": "Todas as soluções",
            "entrada": f"tamanho = {len(vetor)}",
            "qtd_solucoes": len(sols),
            "chamadas": cont["chamadas"],
            "iteracoes": cont["iteracoes"],
            "tempo": tempo
        })

    return pd.DataFrame(resultados)

# EX: 

sol, cont, tempo = n_rainhas_primeira(4)

print("Primeira solução para 4-rainhas:")
mostrar_tabuleiro(sol)

print("\nContadores:")
print(cont)

print("\nTempo:")
print(tempo)

print("\nTabela de resultados:")
df = tabela_resultados()
print(df)
```
